Question 2

```python
# Extract the amenities column
amenities = df['amenities']

# Define a dictionary to store the frequency of each amenity
amenity_count = {}

# Loop through each row of the amenities column
for row in amenities:
    # Split the amenities by comma
    row_amenities = row.strip("{}").replace('"', '').split(",")
    
    # Loop through each amenity in the row
    for amenity in row_amenities:
        # Add the amenity to the dictionary and increment its count
        if amenity in amenity_count:
            amenity_count[amenity] += 1
        else:
            amenity_count[amenity] = 1

# Convert the dictionary to a pandas DataFrame
amenity_df = pd.DataFrame(list(amenity_count.items()), columns=['Amenity', 'Frequency'])

# Sort the DataFrame by frequency in descending order
amenity_df = amenity_df.sort_values('Frequency', ascending=False)

# Display the top 10 amenities
amenity_df.head(10)
```
